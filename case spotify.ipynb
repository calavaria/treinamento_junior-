```python
df = pd.read_csv("spotify.csv")
df = df.dropna()
ano = df['added'].apply(lambda x : x[0:4])
df = df.assign(year = ano)
mes = df["added"].apply(lambda x : x[5:7])
df= df.assign(month = mes)
dia = df["added"].apply(lambda x : x[8:])
df = df.assign(day = dia)
def transformar(x):
    aux1 = x[0:4]
    aux2 = x[5:7]
    aux3 = x[8:]
    y = aux3 + '/' + aux2 + '/' + aux1
    return y
df['added'] = pd.to_datetime(df['added'].apply(transformar), errors = 'ignore', format = '%d/%m/%y')
df
```

```python
sns.jointplot(x = "dur", y = "year", data = df)
```

```python
sns.pairplot(df,hue = "artist type", palette = 'coolwarm')
```

```python
df3 = df[df['artist type'] == 'Solo']
y = df3.groupby('dur').mean()
x = df3.groupby('top year')
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```
